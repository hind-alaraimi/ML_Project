```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.special import expit
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(n_samples=100, 
                           n_features=1, 
                           n_informative=1, 
                           n_redundant=0, 
                           n_clusters_per_class=1, 
                           random_state=42)
```

```python
X.shape, y.shape
```

```python
clf = LogisticRegression(C=10000)
clf.fit(X,y)
```

```python
plt.scatter(X.ravel(), y, color="black")

X_test = np.linspace(X.min(),X.max(),300)
proba = clf.predict_proba(X_test.reshape((-1,1)))[:,1]
plt.plot(X_test,proba,label="Logistic Regression Model", color="red", linewidth=3)
```

```python
OLS = LinearRegression()
OLS.fit(X,y)
```

```python
plt.scatter(X.ravel(),y,label="--",color='black',zorder = 20)

plt.plot(
    X_test,
    OLS.predict(X_test.reshape((-1,1))),
    label = "Linear reg model",
    linewidth = 1
)

plt.axhline(0.5, color=".5")
```

```python
plt.scatter(X.ravel(),y,color='black')
X_test = np.linspace(X.min(),X.max(), 300)
proba = clf.predict_proba(X_test.reshape((-1,1)))[:,1]
plt.plot(X_test,proba,label="Logistic Regression Model",color="red",linewidth=3)

plt.plot(
    X_test,
    OLS.predict(X_test.reshape((-1,1))),
    label = "Linear Regression Model",
    linewidth = 1
)
plt.axhline(0.5,color=".5")
```
