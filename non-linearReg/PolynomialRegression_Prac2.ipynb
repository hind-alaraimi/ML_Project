```python
import numpy as np
from matplotlib import pyplot as plt

from sklearn.linear_model import LinearRegression

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Ridge
```

```python
X = 10 * np.random.rand(100, 1) - 5  # X in [-5, 5]
y = 0.5 * X + np.sin(X) + np.random.randn(100, 1)
```

```python
#X = 6* np.random.rand(100, 1) - 6
#y = 0.3 * X**2 + X + 2 + np.random.randn(100, 1)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(X,y)
```

```python
model = LinearRegression()
model.fit(X,y)
y_hat = model.predict(X)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(X,y)
ax.plot(X,y_hat)
```

```python
X_poly = PolynomialFeatures(3).fit_transform(X)
```

```python
X_poly.shape
```

```python
model2 = Ridge(alpha=0.003)
model2.fit(X_poly,y)
y_hat2 = model2.predict(X_poly)
X_b = np.linspace(X.min(),X.max(),10).reshape(-1,1)
X_b_poly = PolynomialFeatures(3).fit_transform(X_b)
y_b_hat = model2.predict(X_b_poly)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(X,y)
ax.plot(X_b,y_b_hat,color="red")
ax.set_xlabel("x")
ax.set_ylabel("y")
```

### Make pipeline

```python
from sklearn.pipeline import make_pipeline
```

```python
pipeline = make_pipeline(PolynomialFeatures(10),LinearRegression())
```

```python
pipeline.fit(X,y)
```

```python
X_b = np.linspace(X.min(),X.max(),100).reshape(-1,1)
```

```python
y_hat_b = pipeline.predict(X_b)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(X,y)
ax.plot(X_b,y_hat_b,color="red")
ax.set_ylabel("y")
ax.set_xlabel("x")
```
